```python
import pandas as pd
import plotly.graph_objects as go
import random

random.seed()
```

```python
def randCards(nums, per):
    n = 0
    for i in range(nums):
        if random.random() < per:
            n = n + 1
    
    return n

def randCards2(nums, per, maxtimes):
    n = 0
    cn = 0
    for i in range(nums):
        if cn >= maxtimes - 1:
            n = n + 1
            cn = 0            
        elif random.random() < per:
            n = n + 1
            cn = 0
        else:
            cn = cn + 1
    
    return n

def randCards3(nums, per, maxtimes):
    n = 0
    cn = 0
    bd = 0
    for i in range(nums):
        if cn >= maxtimes - 1:
            n = n + 1
            cn = 0            
            bd += 1
        elif random.random() < per:
            n = n + 1
            cn = 0
        else:
            cn = cn + 1
    
    return n, bd
```

```python
ret1 = {}
ret2 = {}
for i in range(1000000):
    rn = randCards(10, 0.1)
    if rn not in ret1:
        ret1[rn] = 1
    else:
        ret1[rn] = ret1[rn] + 1
        
    rn = randCards2(10, 0.1, 10)
    if rn not in ret2:
        ret2[rn] = 1
    else:
        ret2[rn] = ret2[rn] + 1        
```

```python
def showBars(tags, rets, totalnums):
    fig = go.Figure()
    
    for i in range(len(tags)):
        xarr = []
        yarr = []
        
        for k, v in rets[i].items():
            xarr.append(k)
            yarr.append(v / totalnums)
            
        fig.add_trace(go.Bar(
            x=xarr,
            y=yarr,
            name=tags[i],
        ))            

    fig.update_layout(barmode='group', xaxis_tickangle=-45)
    fig.show()
    
def showPie(name, ret, totalnums, funcOnName):
    tagarr = []
    valarr = []

    for k, v in ret.items():
        if funcOnName != None:
            tagarr.append(funcOnName(k))
        else:
            tagarr.append(k)

        valarr.append(v / totalnums)
            
    fig = go.Figure(data=[go.Pie(labels=tagarr, 
                                 values=valarr, 
                                 textinfo='label+percent', 
                                 insidetextorientation='radial')])

    fig.show()
```

```python
showBars(['百万人连抽10次-无保底', '百万人连抽10次-保底'], [ret1, ret2], 1000000)
```

```python
# 平均成本，假设抽1次成本10，得到一张收益100
def randCost(nums, randnums, per, maxtimes):
    cost = 0
    getnums = 0
    for i in range(nums):
        rn = randCards2(randnums, per, maxtimes)
        
        cost = cost + randnums
        getnums = getnums + rn
#         cost = cost + (rn * 100) - (randnums * 10)
    
    return cost, getnums
```

```python
def showBars2(name, xarr, yarr):
    fig = go.Figure()
            
    fig.add_trace(go.Bar(
        x=xarr,
        y=yarr,
        name=name,
    ))            

    fig.update_layout(barmode='group', xaxis_tickangle=-45)
    fig.show()
    
def showBars3(names, xarrs, yarrs):
    fig = go.Figure()
    
    for i in range(len(names)):
        fig.add_trace(go.Bar(
            x=xarrs[i],
            y=yarrs[i],
            name=names[i],
        ))            

    fig.update_layout(barmode='group', xaxis_tickangle=-45)
    fig.show()
```

```python
xarr = []
yarr = []
yarr2 = []

for i in range(20):
    cc, cg = randCost(1000000, i + 1, 0.1, 10)
    xarr.append(i + 1)
    yarr.append(cc)
    yarr2.append(cg)
    
showBars3(['花费', '获得'], [xarr, xarr], [yarr, yarr2])
```

```python
# 最多抽到保底，如果不到保底，抽到第几次就停
def randCardsEx(nums, per, maxtimes):
    n = 0
    cn = 0
    cnn = 0
    tn = 0
    while True is True:
        tn = tn + 1
        
        if cn >= maxtimes - 1:
            n = n + 1
            cn = 0
            
            return n, tn
        elif random.random() < per:
            n = n + 1
            cn = 0
            
            if n >= 2:
                return n, tn
        else:
            cn = cn + 1
            
            if n > 0:
                cnn = cnn + 1
                if cnn >= nums:
                    return n, tn
    
    return n

# 平均成本，假设抽1次成本10，得到一张收益100
def randCostEx(nums, randnums, per, maxtimes):
    cost = 0.0
    for i in range(nums):
        rn, cn = randCardsEx(randnums, per, maxtimes)
#         cost = cost + (rn * 100) - (cn * 10)
        cost = cost + (cn * 10)        
    
    return cost
```

```python
xarr = []
yarr = []

for i in range(10):
    cc = randCostEx(1000000, i, 0.1, 10)
    xarr.append(i + 1)
    yarr.append(cc)
    
showBars2('抽到后再多抽n次回报率', xarr, yarr)
```

```python
# 最多抽到保底，如果不到保底，抽到第几次就停
def randCardsEx2(nums, per, maxtimes):
    n = 0
    cn = 0
    cnn = 0
    tn = 0
    while True is True:        
        if n >= 2:
            if cnn >= nums:
                return n, tn
            
        tn = tn + 1            
        
        if cn >= maxtimes - 1:
            n = n + 1
            cn = 0
            
            if n >= 2:
                return n, tn
        elif random.random() < per:
            n = n + 1
            cn = 0
            
            if n >= 3:
                return n, tn
            
            if n >= 2:
                cnn = cnn + 1            
        else:
            cn = cn + 1
            
            if n >= 2:
                cnn = cnn + 1
    
    return n

# 平均成本，假设抽1次成本10，得到一张收益100
def randCostEx2(nums, randnums, per, maxtimes):
    cost = 0.0
    for i in range(nums):
        rn, cn = randCardsEx2(randnums, per, maxtimes)
        cost = cost + (rn * 100) - (cn * 10)
        cost = cost + (cn * 10)
    
    return cost
```

```python
xarr = []
yarr = []

for i in range(10):
    cc = randCostEx2(1000000, i, 0.1, 10)
    xarr.append(i + 1)
    yarr.append(cc)
    
showBars2('抽到后再多抽n次回报率', xarr, yarr)
```
