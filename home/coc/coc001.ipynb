```python
def procCOCDF(df):
    for i, d in df.iterrows():
        if i == 0:
            continue
        
        if i == 1:
            for cc in df.columns.values:
                if pd.isna(d[cc]):
                    if df.loc[0][cc] == 'String':
                        df.loc[i][cc] = ''
                    elif df.loc[0][cc] == 'boolean':
                        df.loc[i][cc] = False
                    elif df.loc[0][cc] == 'int':
                        df.loc[i][cc] = 0                    
        else:
            for cc in df.columns.values:
                if pd.isna(d[cc]):
                    df.loc[i][cc] = df.loc[i - 1][cc]
                

procCOCDF(df)
df
```
